# Double Slit Interference

Wave equation :  $\boxed{\Psi(x,t) = A\sin(kx-\omega t)}$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.animation as anim
x_l = 20
y_l = 5
points = 1000
x = np.linspace(0,x_l,points)
y = np.linspace(-y_l,y_l,points)
gridx , gridy = np.meshgrid(x,y)
```

# Variables

These are variable definitions below :
1. $d$ is the distance between the two slits
2. $dist\_measure$ is the distance where we plot the intensity map
3. $A$ is the amplitude of the source wave
4. $omg$ is the angular frequency of the wave
5. $k$ is the angular wavenumber of the wave
6. $t$ is where the cmap is plotted(just a reference needed, actual value does not matter)
7. $time$ is the time where the intensity is plotted.

```python
d = 2
# dist_measure = 49.009
dist_measure = 
A = 50
omg = 10
k=10
time = 2*np.pi/(omg)
t=0
dist_y_1 = gridy+d
dist_y_2 = gridy-d
```

```python
dist_1 = np.sqrt(gridx**2+dist_y_1**2)
dist_2 = np.sqrt(gridx**2+dist_y_2**2)
def wave1(t,dist):
    return A * np.sin(k*dist - omg* t)
def wave2(t,dist):
    return A * np.sin(k*dist - omg * t)
def wave(t,dist1,dist2):
    return wave1(t,dist1)+wave2(t,dist2)
```

```python
fig,axes = plt.subplots(figsize = (16,9))
axes.set_xlim(0,x_l-1)
axes.set_ylim(-y_l-0.1,y_l-0.1)
axes.scatter(gridx,gridy,c=wave(0,dist_1,dist_2)**2,cmap = "viridis")
```

```python
gg = np.array([gridy[i][0] for i in range(0,points)])
gx = np.array([dist_measure for i in range(points)])
gx0 = np.array([0 for i in range(points)])
fig1,axes1 = plt.subplots(figsize = (16,9))
axes1.set_xlim(-2,2)
axes1.plot(gg,wave(time,gg,gx)**2,lw=5,label = "dist = %d cm"%dist_measure)
axes1.plot(gg,wave(time,gg,gx0)**2,label = "dist = %d cm"%0)
axes1.legend(loc = "best")
axes1.grid()
```
